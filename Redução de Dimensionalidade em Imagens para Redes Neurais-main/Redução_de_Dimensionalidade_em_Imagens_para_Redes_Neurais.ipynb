Objetivo: Converter uma imagem RGB para tons de cinza, em seguida transformar a imagem em tons de cinza para binária

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread("sample_data/cachorro1.jpg")
imgplot = plt.imshow(img)
plt.show()
```

```python
#Convertendo imagem em tons de cinza
img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
#mostrando imagem em tons de cinza
plt.imshow(img, cmap = plt.get_cmap(name = 'gray'))
```

```python
#Salvando imagem em tons de cinza
mpimg.imsave("cachorro_cinza.jpg", img)
```

```python
#convertendo a imagem para binário
#pixels com valores acima de 170 serão convertidos para 1, o resto fica em 0
cachorro_binario = 1 * (img > 170)

#Mostrando a imagem binária
plt.imshow(cachorro_binario, cmap = plt.get_cmap(name = 'gray'))
```
